```python
# retrieve parts of speech using spacy
#named entity recognition using spacy
#visualize POS and NER
# sentence segmentation


import spacy
nlp = spacy.load('en_core_web_sm')
```

```python
doc = nlp(u"that was really bad and people just left that place as soon as they dead")

for token in doc:
    
    print(f" {token.text:{10}} {token.pos_:{10}} {spacy.explain(token.tag_)}")
```

```python
from spacy import displacy
displacy.render(doc,style = 'dep', options = {'font':'Areal','distance':70
                                              ,'color': 'Blue','bg':'white'}, jupyter =True)
```

```python
def showent(doc):
    if doc.ents:
        for ent in doc.ents:
            print(f"{ent.text:{10}} {ent.label_:{10}} {spacy.explain(ent.label_)}")
    else:
        print("No NER")
```

```python
doc = nlp(u"they are the best people in Accenture and they are from South-East region!")
```
